```python
from sklearn.cluster import MiniBatchKMeans
```

### Task 5: K-means Image Compression with Interactive Controls
---

```python
img_dir = "images/"
```

```python
@interact
def color_compression(image = os.listdir(img_dir), k = IntSlider(min=1, max=256, step=1, value=16, continuous_update = False, layout=dict(width="100%"))):
    input_img = io.imread(img_dir + image)
    img = input_img/255
    img = img.reshape(-1, 3)
    kmeans = MiniBatchKMeans(k)
    kmeans = kmeans.fit(img)
    k_colors = kmeans.cluster_centers_[kmeans.predict(img)]
    
    img = np.reshape(k_colors, (input_img.shape))
    fig, (ax1, ax2) = plt.subplots(1,2)
    fig.suptitle("K-Means Image Compression", fontsize=20)
    
    ax1.set_title("Compressed Image")
    ax1.set_xticks([])
    ax1.set_yticks([])
    ax1.imshow(img)
    
    ax2.set_title("Original Image")
    ax2.set_xticks([])
    ax2.set_yticks([])
    ax2.imshow(input_img)
    
    plt.subplots_adjust(top=0.85)
    plt.show()
```
